### Constant Rule
If $c$ is a real number, then $\frac{d}{dx}(c) = 0$

### Power Rule
If $n$ is a postive integer, then $\frac{d}{dx}(x^{n}) = nx^{n-1}$

##### Definitions
$f(x) = x^{n} \implies f(x) - f(a) = x^{n} - a^{n}$

$x^{n} - a^{n} = (x - a)(x^{n-1} + x^{n-2}a + ... + xa^{n-2} + a^{n-1})$

$\lim_{x \rightarrow a} (x^{n-1} + x^{n-2}a + ... + xa^{n-2} + a^{n-1}) = a^{n-1} + a^{n-2} \cdot a + ... + a \cdot a^{n-2} + a^{n-1} = na^{n-1}$

##### Definition Examples
$x^{4} - a^{4} = (x - a)(x^{3} + x^{2}a + xa^{2} + a^{3})$

$\lim_{x \rightarrow a} (x^{3} + x^{2}a + xa^{2} + a^{3}) = a^{3} + a^{2} \cdot a + a \cdot a^{2} = a^{3} + a^{3} + a^{3} = 3n^{3}$

##### Proof
$$
\begin{align}
f'(a) &= \lim_{x \rightarrow a} \frac{f(x) - f(a)}{x - a} \\
&= \lim_{x \rightarrow a} \frac{x^{n} - a^{n}}{x - a} \\
&= \lim_{x \rightarrow a} \frac{(x - a)(x^{n-1} + x^{n-2}a + ... + xa^{n-2} + a^{n-1})}{x - a} \\
&= \lim_{x \rightarrow a} (x^{n-1} + x^{n-2}a + ... + xa^{n-2} + a^{n-1}) \\
&= a^{n-1} + a^{n-2} \cdot a + ... + a \cdot a^{n-2} + a^{n-1} \\
&= na^{a-1}
\end{align}
$$

### The derivative of $e^{x}$
The function $f(x) = e^{x}$ is differentiable for all real numbers $x$ and
$$\frac{d}{dx}(e^{x}) = e^{x}$$

##### Proof
$$
\begin{align}
\frac{d}{dx}(e^{x}) &= \lim_{h \rightarrow 0} \frac{e^{x + h} - e^{x}}{h} \\
&= \lim_{h \rightarrow 0} \frac{e^{x} \cdot e^{h} - e^{x}}{h} \\
&= \lim_{h \rightarrow 0} \frac{e^{x}(e^{h} - 1)}{h} \\ 
&= e^{x} \cdot \lim_{h \rightarrow 0} \frac{e^{h} - 1}{h} \\
&= e^{x} \cdot 1 = e^{x}
\end{align}
$$

### Product Rule
If $f$ and $g$ are differentiable at $x$, then
$$\frac{d}{dx}[f(x)g(x)] = f'(x)g(x) + f(x)g'(x)$$

##### Definitions
$Z: -f(x)g(x+h) + f(x)g(x + h) = 0$  

$\lim_{h \rightarrow 0} f(x) = f(x)$

$\lim_{h \rightarrow 0} \frac{f(x + h) - f(x)}{h} = f'(x)$

$\lim_{h \rightarrow 0} g(x + h) = g(x)$

$\lim_{h \rightarrow 0} \frac{g(x + h) - g(x)}{h} = g'(x)$

##### Proof
$$
\begin{align}
\frac{d}{dx}[f(x)g(x)] &= \lim_{h \rightarrow 0} \frac{f(x + h)g(x + h) - f(x)g(x)}{h} \\
&= \lim_{h \rightarrow 0} \frac{f(x + h)g(x + h) - f(x)g(x) + Z}{h} \\
&= \lim_{h \rightarrow 0} \frac{f(x + h)g(x + h) - f(x)g(x) + [-f(x)g(x+h) + f(x)g(x + h)]}{h} \\
&= \lim_{h \rightarrow 0} \frac{f(x + h)g(x + h) - f(x)g(x+h) + f(x)g(x + h) - f(x)g(x)}{h} \\
&= \lim_{h \rightarrow 0} \frac{f(x + h)g(x + h) - f(x)g(x+h)}{h} + \lim_{h \rightarrow 0} \frac{f(x)g(x + h) - f(x)g(x)}{h} \\
&= \lim_{h \rightarrow 0} \bigg[\frac{f(x + h) - f(x)}{h} \cdot g(x + h)\bigg] + \lim_{h \rightarrow 0} \bigg[f(x) \cdot \frac{g(x + h) - g(x)}{h} \bigg] \\
&= f'(x) \cdot g(x) + f(x) \cdot g'(x)
\end{align}
$$

### Quotient Rule
If $f$ and $q$ are differentiable at $x$, then the derivative of $f/g$ at $x$ exists provided $g(x) \neq 0$ and

$$\frac{d}{dx}\bigg[\frac{f(x)}{g(x)}\bigg] = \frac{g(x)f'(x) - f(x)g'(x)}{[g(x)]^{2}}$$

##### Definitions
$q(x) = \frac{f(x)}{g(x)}$

$f(x) = g(x)q(x)$

##### Calculation
If $\frac{d}{dx}f/g$ exists,
$$
\begin{align}
f'(x) &= g'(x)q(x) + g(x)q'(x) \\
q'(x) &= \frac{f'(x) - g'(x)q(x)}{g(x)} \\
q'(x) &= \frac{f'(x) - g'(x)\frac{f(x)}{g(x)}}{g(x)} \\
&= \frac{g(x) \Big(f'(x) - g'(x)\frac{f(x)}{g(x)}\Big)}{g(x) \cdot g(x)} \\
&= \frac{g(x)f'(x) - f(x)g'(x)}{[g(x)]^{2}}
\end{align}
$$

##### Definitions
Let $F = f/g$ be the quotient of two functions that are differentiable at $x$

$Z: -f(x)g(x) + f(x)g(x) = 0$  

$\lim_{h \rightarrow 0} f(x) = f(x)$

$\lim_{h \rightarrow 0} \frac{f(x + h) - f(x)}{h} = f'(x)$

$\lim_{h \rightarrow 0} g(x) = g(x)$

$\lim_{h \rightarrow 0} \frac{g(x) - g(x + h)}{h} = -g'(x)$

$\lim_{h \rightarrow 0} \frac{1}{g(x)g(x + h)} = \frac{1}{[g(x)]^{2}}$

##### Proof
$$
\begin{align}
\frac{d}{dx}[f(x)g(x)] &= \lim_{h \rightarrow 0} \frac{\frac{f(x + h)}{g(x + h)} - \frac{f(x)}{g(x)}}{h} \\
&= \lim_{h \rightarrow 0} \frac{f(x + h)}{g(x + h) \cdot h} - \frac{f(x)}{g(x) \cdot h} \\
&= \lim_{h \rightarrow 0} \frac{f(x + h)g(x) - f(x)g(x + h)}{g(x)g(x + h) \cdot h} \\
&= \lim_{h \rightarrow 0} \frac{f(x + h)g(x) - f(x)g(x + h) + Z}{g(x)g(x + h) \cdot h} = \lim_{h \rightarrow 0} \frac{f(x + h)g(x) - f(x)g(x + h) + [-f(x)g(x) + f(x)g(x)]}{g(x)g(x + h) \cdot h} \\
&= \lim_{h \rightarrow 0} \frac{f(x + h)g(x) - f(x)g(x) + g(x)f(x) - g(x+h)f(x)}{g(x)g(x + h) \cdot h} \\
&= \lim_{h \rightarrow 0} \frac{f(x + h)g(x) - f(x)g(x)}{g(x)g(x + h) \cdot h} + \frac{g(x)f(x) - g(x + h)f(x)}{g(x)g(x + h) \cdot h} \\
&= \bigg[\lim_{h \rightarrow 0} \frac{f(x + h) - f(x)}{g(x)g(x + h) \cdot h} \cdot g(x) \bigg] + \bigg[\lim_{h \rightarrow 0} \frac{g(x) - g(x + h)}{g(x)g(x + h) \cdot h} \cdot f(x)\bigg] \\
&= \bigg[\lim_{h \rightarrow 0} \frac{f(x + h) - f(x)}{h} \cdot g(x) \cdot \frac{1}{g(x)g(x + h)} \bigg] + \bigg[\lim_{h \rightarrow 0} \frac{g(x) - g(x + h)}{h} \cdot f(x)  \cdot \frac{1}{g(x)g(x + h)} \bigg] \\
&= \bigg[f'(x) \cdot g(x) \cdot \frac{1}{[g(x)]^{2}} \bigg] + \bigg[-g'(x) \cdot f(x)  \cdot \frac{1}{[g(x)]^{2}} \bigg] \\
&= \frac{f'(x)g(x) - g'(x)f(x)}{[g(x)]^{2}}
\end{align}
$$

### The derivative of $e^{kx}$
For real numbers $k$, 
$$\frac{d}{dx}(e^{kx}) = ke^{kx}$$

##### Example
$$
\begin{align}
\frac{d}{dx}(e^{2x}) &= \frac{d}{dx}(e^{x} \cdot e^{x}) \\
&= \frac{d}{dx}(e^{x}) \cdot e^{x} + e^{x} \cdot \frac{d}{dx}(e^{x}) \\
&= e^{x} \cdot e^{x} + e^{x} \cdot e^{x} = 2e^{2x}
\end{align}
$$

##### Proof by induction for positive integers $k$
Base case ($k = 1$)
$$
\begin{align}
\frac{d}{dx}e^{kx} &= \frac{d}{dx}e^{1 \cdot x} \\
&= \frac{d}{dx}e^{x} \\
&= e^{x}
\end{align}
$$

Inductive step (inductive hypothesis: assume true for $k = n$)
$$
\begin{align}
\frac{d}{dx}e^{nx} = ne^{nx}
\end{align}
$$

Prove true for $k = n + 1$: $\frac{d}{dx}e^{(n+1)x} = (n+1)e^{(n+1)x}$

$$
\begin{align}
\frac{d}{dx}e^{(n+1)x} &= \frac{d}{dx}e^{nx + x} \\
&= \frac{d}{dx}(e^{nx} \cdot e^{x}) \\
&= ne^{nx} \cdot e^{x} + e^{nx} \cdot e^{x} \\
&= e^{x}(ne^{nx} + e^{nx}) \\
&= e^{x} \cdot e^{nx}(n + 1) \\
&= e^{nx+x}(n+1) \\
&= e^{(n+1)x}(n+1) \\
&= (n+1)e^{(n+1)x}
\end{align}
$$

##### Proof by induction for negative integers $k$
Base case ($k = -1$)
$$
\begin{align}
\frac{d}{dx}e^{kx} &= \frac{d}{dx}e^{-1 \cdot x} \\
&= \frac{d}{dx}e^{-x} = \frac{d}{dx}\bigg(\frac{1}{e^{x}}\bigg) \\
&= \frac{o \cdot e^{x} - 1 \cdot e^{x}}{e^{x} \cdot e^{x}} = \frac{-e^{x}}{[e^{x}]^{2}} = \frac{-1}{e^{x}} = -e^{-x}
\end{align}
$$

Inductive step (inductive hypothesis: assume true for $k = -n$)
$$
\begin{align}
\frac{d}{dx}e^{-nx} = -ne^{-nx}
\end{align}
$$

Prove true for $k = -n - 1$: $\frac{d}{dx}e^{(-n - 1)x} = (-n - 1)e^{(-n - 1)x}$

$$
\begin{align}
\frac{d}{dx}e^{(-n - 1)x} &= \frac{d}{dx}e^{-nx - x} \\
&= \frac{d}{dx}(e^{-nx} \cdot e^{-x}) \\
&= \frac{d}{dx}\bigg(\frac{e^{-nx}}{e^{x}}\bigg) \\
&= \frac{-ne^{-nx} \cdot e^{x} - e^{-nx} \cdot e^{x}}{e^{x} \cdot e^{x}} = \frac{e^{x}(-ne^{-nx} - e^{-nx})}{e^{x} \cdot e^{x}} = \frac{-ne^{-nx} - e^{-nx}}{e^{x}} = \frac{e^{-nx}(-n - 1)}{e^{x}} \\
&= e^{-nx} \cdot e^{-x} (-n - 1) = e^{-nx-x}(-n - 1) = e^{(-n - 1)x}(-n - 1) = (-n - 1)e^{(-n - 1)x} 
\end{align}
$$
